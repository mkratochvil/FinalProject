# Background

The objective function right now is for including the first stage variables in the second stage. In particular, define $S$ as the set of scenarios, $c$ and $x$ as the first stage costs and variables, and $d$ and $y$ as the second stage costs and variables. Note that $d$ gives the costs for daily operations, and we are looking to minimize over some target year. We have the objective function $f(x,y)$ defined as:

$$ f(x,y) = c^Tx + \frac{365.25}{|S|}\sum_{s\in S} d^T y_s  = \frac{1}{|S|} \sum_{s\in S} ( c^Tx + 365.25\cdot d^Ty )$$

Since our power generation variables are generally divided by a base power of 100, we need to make sure to scale our costs by multiplying 100 to whatever variable appears here (for both $c$ and $d$). 

```julia
using Pkg
Pkg.activate(".")
```

```julia
using JuMP
using DataFrames
using CSV
```

```julia
include("./parameters.jl")
include("./get_functions.jl")
```

```julia
model = JuMP.read_from_file("./new_store_exp_Z1_newptdf.mps")
```

```julia
objfunc = JuMP.objective_function(model);
```

```julia
objfuncdict = objfunc.terms;
```

```julia
count = 0
for (key, value) in objfuncdict
    count += 1
    println("$(count), $(key), $(value)")
end
```

```julia
# adjust thermal for average year (not just leap year)
count = 0
for gen in gens
    for ts in timesteps
        count+=1
        var = get_thermal_variable(model,gen,ts)
        oldcost = objfuncdict[var]
        JuMP.set_objective_coefficient(model, var, 365.25/366*oldcost)
        println(count, " ", var, " ", " ", oldcost, " ", JuMP.objective_function(model).terms[var])
    end
end
```

```julia
# adjust loss of load to 2000/MWh in day for average year (not just leap year)
count = 0
for bus in buses
    for ts in timesteps
        count+=1
        var = get_lossofload_variable(model,bus,ts)
        oldcost = objfuncdict[var]
        JuMP.set_objective_coefficient(model, var, 0.5*365.25/366*oldcost)
        println(count, " ", var, " ", " ", oldcost, " ", JuMP.objective_function(model).terms[var])
    end
end
```

```julia
# adjust overload to 2000/MWh in day for average year (not just leap year)
count = 0
for bus in buses
    for ts in timesteps
        count+=1
        var = get_overload_variable(model,bus,ts)
        oldcost = objfuncdict[var]
        JuMP.set_objective_coefficient(model, var, 0.5*10*365.25/366*oldcost)
        println(count, " ", var, " ", " ", oldcost, " ", JuMP.objective_function(model).terms[var])
    end
end
```

```julia
# set charging operational cost to 1 per MW (remember this is hourly)
count = 0
for bus in buses
    for ts in timesteps
        count+=1
        var = get_charging_variable(model,bus,ts)
        #oldcost = objfuncdict[var]
        JuMP.set_objective_coefficient(model, var, 365.25)
        println(count, " ", var, " ", " ", " ", JuMP.objective_function(model).terms[var])
    end
end
```

```julia
# set discharging operational cost to 18 per MW (remember this is hourly)
count = 0
for bus in buses
    for ts in timesteps
        count+=1
        var = get_discharging_variable(model,bus,ts)
        #oldcost = objfuncdict[var]
        JuMP.set_objective_coefficient(model, var, 365.25*18)
        println(count, " ", var, " ", " ", " ", JuMP.objective_function(model).terms[var])
    end
end
```

```julia
#check everything is correct
count = 0
for (key, value) in JuMP.objective_function(model).terms
    count+=1
    println(count, " ", key, " ", value)
end
```

```julia
# set objective value to 100 times to correspond with base power = 100
count = 0
for (key, valueold) in JuMP.objective_function(model).terms
    count+=1
    JuMP.set_objective_coefficient(model, key, 100*valueold)
    println(count, " ", key, " ", valueold, " ", JuMP.objective_function(model).terms[key])
end
```

```julia
# set objective value to 1e-6 times to represent final cost in millions
count = 0
for (key, valueold) in JuMP.objective_function(model).terms
    count+=1
    JuMP.set_objective_coefficient(model, key, 1e-6*valueold)
    println(count, " ", key, " ", valueold, " ", JuMP.objective_function(model).terms[key])
end
```
