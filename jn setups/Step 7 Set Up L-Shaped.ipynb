```julia
using JuMP
using DataFrames
using CSV
using Gurobi
using DataFrames
```

```julia
include("./parameters.jl")
include("./get_functions.jl")

loadcsv = CSV.File("./LOAD.csv");
include("./modification_functions.jl")
loaddis = load_distribution_dict(loadcsv);
ptdfdict = ptdf_dict(ptdfdf);
```

```julia
nscen = 12
```

```julia
loaddf = DataFrame(CSV.File("./hourly_load_12scen.csv"))
base = 100.0;
factor = 1.35;

load = Dict()

for i = 1:nscen
    load[i] = [];
    for j = 1:24
        push!(load[i], factor/base*loaddf[j,i])
    end
end
```

```julia
winddf = DataFrame(CSV.File("./hourly_wind122_12scen.csv"))
base = 100.0;
factor = 1.0;

wind = Dict()

for i = 1:nscen
    wind[i] = [];
    for j = 1:24
        push!(wind[i], factor/base*winddf[j,i])
    end
end
```

```julia
model = JuMP.read_from_file("./sub_12scen_test.mps")
```

```julia
#fixing energy to be 0 at start and end of day
for bus in buses

    JuMP.fix(get_stored_variable(model,bus,1), 0.0)
    JuMP.fix(get_stored_variable(model,bus,24), 0.0)
end
```

```julia
for (var, coeff) in JuMP.objective_function(model).terms
    println(var, " ", coeff)
end
```

```julia
JuMP.set_optimizer(model, Gurobi.Optimizer)
```

```julia
JuMP.optimize!(model)
```

```julia
get_branch_interval(model,"A3",1)
```

```julia
#objective for 12th scenario should be 2.066154697e+02
for scen = 1:nscen

    println("...Scenario $(scen)...")

    loadvec = load[scen]
    windvec = wind[scen]

    println("......Adjusting uncertainty...")
    println(".........load balance...")
    adjust_load_balance_constraint!(model, loadvec, loaddis)  
    println(".........ptdf...")
    adjust_ptdf_constraint!(model, model, ptdfdict, loadvec, loaddis) 
    println(".........wind upper bound...")
    adjust_wind_ub!(model, windvec)
    
    JuMP.write_to_file(model, "./submodels/scen_$(scen).mps")
end
```

```julia
# prototype first stage model function would start here

test1model = JuMP.read_from_file("./submodels/scen_1.mps")
```

```julia
for ts in timesteps
    
    for gen in gens
        JuMP.delete(test1model,get_thermal_variable(test1model,gen,ts))
    end
    for br in branches
        JuMP.delete(test1model,get_line_variable(test1model,br,ts))
    end
    
    JuMP.delete(test1model, get_wind_variable(test1model,122,ts))
    
    for bus in buses
        JuMP.delete(test1model, get_charging_variable(test1model,bus,ts))
        JuMP.delete(test1model, get_discharging_variable(test1model,bus,ts))
        JuMP.delete(test1model, get_stored_variable(test1model,bus,ts))
        JuMP.delete(test1model, get_lossofload_variable(test1model,bus,ts))
        JuMP.delete(test1model, get_overload_variable(test1model,bus,ts))
    end
end
```

```julia
for ts in timesteps
    
    for gen in gens
        if ts > 1
            JuMP.delete(test1model, get_ramp_lb(test1model, gen,ts))
            JuMP.delete(test1model, get_ramp_ub(test1model, gen,ts))
        end
        JuMP.delete(test1model, get_thermal_interval(test1model, gen,ts))
    end
    
    for br in branches
        JuMP.delete(test1model, get_ptdf_con(test1model, br,ts))
        JuMP.delete(test1model, get_branch_interval(test1model, br,ts))
    end
    
    for bus in buses
        JuMP.delete(test1model, get_storage_balance(test1model, bus,ts))
        JuMP.delete(test1model, get_load_balance(test1model, bus,ts))
        if bus == 122
            JuMP.delete(test1model, get_wind_lb(test1model, bus,ts))
            JuMP.delete(test1model, get_wind_ub(test1model, bus,ts))
        end
        JuMP.delete(test1model, get_charge_lb(test1model, bus,ts))
        JuMP.delete(test1model, get_discharge_lb(test1model, bus,ts))
        JuMP.delete(test1model, get_storage_lb(test1model, bus,ts))
        JuMP.delete(test1model, get_charge_ub(test1model, bus,ts))
        JuMP.delete(test1model, get_discharge_ub(test1model, bus,ts))
        JuMP.delete(test1model, get_storage_ub(test1model, bus,ts))
        JuMP.delete(test1model, get_loss_of_load_lb(test1model, bus,ts))
        JuMP.delete(test1model, get_overload_lb(test1model, bus,ts))
    end
end
```

```julia
vardict = Dict{Int64,Array{Any}}()
```

```julia
modelvars = JuMP.read_from_file("./storage_expansion/second_stage/scen_1.mps")
```

```julia
firstvars = []
for bus in buses
    push!(firstvars,(string(get_PR_variable(modelvars,bus)), 0.0, Inf, 0.0))
    push!(firstvars,(string(get_ER_variable(modelvars,bus)), 0.0, Inf, 0.0))
end

vardict[1] = firstvars
```

```julia
secondvars = [];

for ts in timesteps
    
    for gen in gens
        push!(secondvars,string(get_thermal_variable(modelvars,gen,ts)))
    end
    for br in branches
        push!(secondvars,string(get_line_variable(modelvars,br,ts)))
    end
    
    push!(secondvars, string(get_wind_variable(modelvars,122,ts)))
    
    for bus in buses
        push!(secondvars, string(get_charging_variable(modelvars,bus,ts)))
        push!(secondvars, string(get_discharging_variable(modelvars,bus,ts)))
        push!(secondvars, string(get_stored_variable(modelvars,bus,ts)))
        push!(secondvars, string(get_lossofload_variable(modelvars,bus,ts)))
        push!(secondvars, string(get_overload_variable(modelvars,bus,ts)))
    end
end

vardict[2] = secondvars
```

```julia
function first_func()
    
    model = JuMP.read_from_file("./storage_expansion/first_stage/first_stage_model.mps")
    JuMP.set_optimizer(model, Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0) 
    
    return model
end
```

```julia
function second_func(scen)
    
    model = JuMP.read_from_file("./storage_expansion/second_stage/scen_$(scen).mps")
    JuMP.set_optimizer(model, Gurobi.Optimizer)
    set_optimizer_attribute(model, "OutputFlag", 0) 
    return model
end
```

```julia
x, firststage, fs, ittime, niter = LShaped.L_Shaped_Algorithm_new(second_func, 
                            vardict, 
                            12, 
                            first_func, 
                            1e-6, 
                            100);
```
