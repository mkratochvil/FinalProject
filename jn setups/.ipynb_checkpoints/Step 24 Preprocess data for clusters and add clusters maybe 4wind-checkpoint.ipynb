```julia
X1458 = Array{Float64}(undef, size(X,1), 1458)

for exp = 1:5
    
    trvec = collect(part1458[exp,:])
    
    for i = 1:1458
        X1458[:,i] = X[:, trvec[i]]
    end

    for K in [12,24]
        Random.seed!(7)
        C = Clustering.kmeans(X1458,K)
        
        cldict = Dict()
        for i = 1:K
            cldict[i] = Int64[]
        end
        for i = 1:1458
            push!(cldict[C.assignments[i]], trvec[i])
        end
        
        lowest_cost = Inf*ones(K)
        closest_center = zeros(Int64,1,K)

        for i = 1:1458
            if C.costs[i] < lowest_cost[C.assignments[i]]
                closest_center[1,C.assignments[i]] = trvec[i]
                lowest_cost[C.assignments[i]] = C.costs[i]
            end
        end
        
        CSV.write("../scenarios/four_wind/cldict_$(K)_$(exp).csv",cldict)
        CSV.write("../scenarios/four_wind/centers_$(K)_$(exp).csv", DataFrame(C.centers))
        CSV.write("../scenarios/four_wind/closest_to_center_$(K)_$(exp).csv", DataFrame(closest_center))
        
        println(K, " ", exp)
        show(C.counts)
        println(closest_center)
        println()
    end
end
        
        
        
    
```
